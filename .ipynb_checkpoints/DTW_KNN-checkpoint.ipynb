```python
import os
import pandas as pd
import numpy as np
from collections import deque
```

```python
# --- Parameters ---
WINDOW_SIZE = 6 # rolling window for Z energy (~50ms)
Z_ENERGY_THRESHOLD = 3.0 # threshold for tap detection
GYRO_THRESHOLD = 1.0 # ignore big rotations
DEBOUNCE_MS = 120 # cooldown between taps
PRE_MS = 60 # pre-window in ms
POST_MS = 200 # post-window in ms
SAMPLING_RATE = 120 # Hz (approximate)
PRE_SAMPLES = int(PRE_MS / 1000 * SAMPLING_RATE)
POST_SAMPLES = int(POST_MS / 1000 * SAMPLING_RATE)
```

```python
# Path to the trial csv files
DATA_FOLDER = "data/"
```

```python
# Output storage
all_segments = []
```

```python
# Map files to labels (grid0–grid8)
def extract_segments_from_folder(folder_path):
    file_list = sorted([f for f in os.listdir(folder_path) if f.endswith('.csv')])
    file_to_label = {f: i for i, f in enumerate(file_list)}

    for file_name in file_list:
        df = pd.read_csv(os.path.join(folder_path, file_name))
        n = len(df)

        highZ = df['highZ'].values
        gx, gy, gz = df['gx'].values, df['gy'].values, df['gz'].values
        gyroMag = np.sqrt(gx**2 + gy**2 + gz**2)

        energy_window = deque(maxlen=WINDOW_SIZE)
        last_tap_idx = -DEBOUNCE_MS * SAMPLING_RATE // 1000

        for i in range(n):
            energy_window.append(highZ[i])
            energyZ = np.mean(np.array(energy_window)**2)

            if (energyZ > Z_ENERGY_THRESHOLD) and (gyroMag[i] < GYRO_THRESHOLD) and (i - last_tap_idx > DEBOUNCE_MS * SAMPLING_RATE // 1000):
                tap_center = i
                start_idx = max(0, tap_center - PRE_SAMPLES)
                end_idx = min(n, tap_center + POST_SAMPLES)

                segment = df.iloc[start_idx:end_idx][['ax','ay','az','highX','highY','highZ','accelMag','gx','gy','gz']].to_numpy()
                all_segments.append({
                    'label': file_to_label[file_name], 
                    'segment': segment})

                last_tap_idx = i

    return all_segments
```

```python
participants = ["data/participant2"]

participant_segments = {}

for p in participants:
    participant_segments[p] = extract_segments_from_folder(p)
    print(p, "segments:", len(participant_segments[p]))
```

```python
all_segments = []

for p in participants:
    all_segments.extend(participant_segments[p])

print("Total segments across all participants:", len(all_segments))
#all participants
```

# Build 1D / 4D / 10D Feature Sequences for DTW + KNN


Now that we have all tap segments extracted (all 10 columns), we create datasets for three feature sets:


- **1D**: `accelMag` only
- **4D**: `[ax, ay, az, accelMag]`
- **10D**: all 10 columns


These datasets can be used directly with DTW + KNN.

```python
# Function to normalize a single tap segment
def normalize_segment(segment, method='zscore'):
    seg = segment.copy()
    if method == 'zscore':
        for i in range(seg.shape[1]):
            mu = np.mean(seg[:, i])
            sigma = np.std(seg[:, i])
            if sigma > 0:
                seg[:, i] = (seg[:, i] - mu) / sigma
    elif method == 'minmax':
        for i in range(seg.shape[1]):
            min_val = np.min(seg[:, i])
            max_val = np.max(seg[:, i])
            if max_val > min_val:
                seg[:, i] = (seg[:, i] - min_val) / (max_val - min_val)
    elif method == 'maxabs':
        for i in range(seg.shape[1]):
            max_val = np.max(np.abs(seg[:, i]))
            if max_val > 0:
                seg[:, i] = seg[:, i] / max_val
    return seg
```

```python
# Initialize datasets
X_1D = []
X_4D = []
X_10D = []
Y = [] # labels


for seg in all_segments:
    data = seg['segment'] # ndarray shape: [timesteps, 10]
    label = seg['label']
    
    
    # Normalize the segment (z-score by default)
    data_norm = normalize_segment(data, method='zscore')
    
    
    # 1D
    X_1D.append(data_norm[:, 6].reshape(-1, 1)) # accelMag
    
    
    # 4D
    X_4D.append(data_norm[:, [0,1,2,6]]) # ax, ay, az, accelMag
    
    
    # 10D
    X_10D.append(data_norm) # all columns
    
    
    # Labels
    Y.append(label)
```

```python
# Convert to numpy arrays (lists of arrays for DTW, keep variable lengths)
import numpy as np
X_1D = np.array(X_1D, dtype=object)
X_4D = np.array(X_4D, dtype=object)
X_10D = np.array(X_10D, dtype=object)
Y = np.array(Y)


print(f"1D dataset: {len(X_1D)} sequences")
print(f"4D dataset: {len(X_4D)} sequences")
print(f"10D dataset: {len(X_10D)} sequences")
print(f"Labels: {np.unique(Y)}")
```

# DTW + KNN for Tap Classification

```python
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from tqdm import tqdm
```

```python
# Split data (70% train, 30% test)
X_train_1D, X_test_1D, y_train, y_test = train_test_split(
    X_1D, Y, test_size=0.3, random_state=42, stratify=Y
)

X_train_4D, X_test_4D, _, _ = train_test_split(
    X_4D, Y, test_size=0.3, random_state=42, stratify=Y
)

X_train_10D, X_test_10D, _, _ = train_test_split(
    X_10D, Y, test_size=0.3, random_state=42, stratify=Y
)

print(f"Training samples: {len(X_train_1D)}, Testing samples: {len(X_test_1D)}")
```

```python
def compute_dtw_distance_matrix(X1, X2):
    """
    Compute DTW distance between each sequence in X1 and X2.
    Returns a distance matrix of shape (len(X1), len(X2))
    """
    distance_matrix = np.zeros((len(X1), len(X2)))
    
    for i, seq1 in enumerate(tqdm(X1, desc="Computing DTW distances")):
        for j, seq2 in enumerate(X2):
            dist, _ = fastdtw(seq1, seq2, dist=euclidean)
            distance_matrix[i, j] = dist
    
    return distance_matrix
```

```python
# Convert labels to integers
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
```

```python
# 1D Example

# Compute training distance matrix (train vs train)
distance_train_1D = compute_dtw_distance_matrix(X_train_1D, X_train_1D)

# Train KNN (metric='precomputed')
knn_1D = KNeighborsClassifier(n_neighbors=1, metric='precomputed')
knn_1D.fit(distance_train_1D, y_train_enc)
```

```python
# Compute test distance matrix (test vs train)
distance_test_1D = compute_dtw_distance_matrix(X_test_1D, X_train_1D)

# Predict
y_pred_1D = knn_1D.predict(distance_test_1D)

# Accuracy
accuracy_1D = np.mean(y_pred_1D == y_test_enc)
print(f"1D DTW + KNN Accuracy: {accuracy_1D*100:.2f}%")
```

```python
# 4D
distance_train_4D = compute_dtw_distance_matrix(X_train_4D, X_train_4D)
knn_4D = KNeighborsClassifier(n_neighbors=1, metric='precomputed')
knn_4D.fit(distance_train_4D, y_train_enc)

distance_test_4D = compute_dtw_distance_matrix(X_test_4D, X_train_4D)
y_pred_4D = knn_4D.predict(distance_test_4D)
accuracy_4D = np.mean(y_pred_4D == y_test_enc)
print(f"4D DTW + KNN Accuracy: {accuracy_4D*100:.2f}%")

# 10D
distance_train_10D = compute_dtw_distance_matrix(X_train_10D, X_train_10D)
knn_10D = KNeighborsClassifier(n_neighbors=1, metric='precomputed')
knn_10D.fit(distance_train_10D, y_train_enc)

distance_test_10D = compute_dtw_distance_matrix(X_test_10D, X_train_10D)
y_pred_10D = knn_10D.predict(distance_test_10D)
accuracy_10D = np.mean(y_pred_10D == y_test_enc)
print(f"10D DTW + KNN Accuracy: {accuracy_10D*100:.2f}%")
```

# Evaluate & Analyze

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
import numpy as np

# Compute confusion matrix
cm = confusion_matrix(y_test_enc, y_pred_10D)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=le.classes_)
fig, ax = plt.subplots(figsize=(6,6))
im = disp.plot(cmap='Blues', ax=ax, values_format='d').im_

ax.set_xlabel("Predicted Label")
ax.set_ylabel("True Label")
ax.set_title("Confusion Matrix with Per-Class Accuracy")

# Add colorbar legend
fig.colorbar(im, ax=ax)

# Compute per-class accuracy (recall per class)
class_accuracy = cm.diagonal() / cm.sum(axis=1)

# Print per-class accuracy
for idx, acc in enumerate(class_accuracy):
    print(f"Class {le.classes_[idx]} accuracy: {acc*100:.2f}%")

# Optional: full classification report (precision, recall, F1)
print("\nDetailed classification report:")
print(classification_report(y_test_enc, y_pred_10D, target_names=[str(c) for c in le.classes_]))
```

```python
acc = cm.diagonal() / cm.sum(axis=1)
plt.bar(le.classes_, acc*100)
plt.xlabel("Grid Cell")
plt.ylabel("Accuracy (%)")
plt.title("Per-Class Accuracy")
plt.show()
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    knn_10D, distance_train_10D, y_train_enc,
    cv=5, train_sizes=np.linspace(0.1, 1.0, 5)
)

train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)

plt.plot(train_sizes, train_mean, 'o-', label='Train')
plt.plot(train_sizes, test_mean, 'o-', label='Test')
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.title('Learning Curve (10D DTW + KNN)')
plt.legend()
plt.show()
```
